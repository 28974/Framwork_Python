```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load all files in the data  folder and merge all into one dataframe df
df1 = pd.read_csv("cfb14.csv", delimiter=",")
df2 = pd.read_csv("cfb15.csv", delimiter=",")
df3 = pd.read_csv("cfb16.csv", delimiter=",")
df4 = pd.read_csv("cfb17.csv", delimiter=",")
df5 = pd.read_csv("cfb18.csv", delimiter=",")
df6 = pd.read_csv("cfb19.csv", delimiter=",")
df7 = pd.read_csv("cfb20.csv", delimiter=",")
df8 = pd.read_csv("cfb21.csv", delimiter=",")
```

```python
college_team_football= pd.concat([df1, df2, df3,df4,df5,df6,df7,df8], ignore_index= True)

college_team_football                 
```

```python
# Add a new column in df, percentage_victory= (Win/games)*100 
college_team_football["percentage_victory"]= (college_team_football["Win"] / college_team_football["Games"])*100
college_team_football

```

```python
# drop the newly created column
new_college_team_football=college_team_football.drop("percentage_victory", axis=1)
new_college_team_football
```

```python
# Look at the documentation of pandas.corr() function and using to compute the matrix of correlation (store this in cor variable)
# Answer
cor = new_college_team_football.corr()
```

```python
cor
```

```python
#Correlation with respect to Win variable and filtering out variables with less correlation.
columns = cor[cor["Win"]>0.75]['Win']
columns_list = columns.index
columns_list

# Explain this code
```

*Ici on reer une serie contenant les colonnes du dataframe correlees a plus de0.75 avec la colonne "Win" et il nous renvoie la liste des index stoke dans la variable "column list"*

```python
# create a dataframe df1: using (df, columns_list) as  data and columns parameters of pandas.Dataframe()
df1 = pd.DataFrame(data=new_college_team_football, columns = columns_list)
df1
```

```python
#Visualization correlation matrix of df2 with the help of seaborn.heatmap
plt.figure(figsize=(12,5))
corre = df1.corr()
heatmap =sns.heatmap(corre, annot= True)
heatmap
```

```python
# Plot colored by continent for years 2000-2007
sns.pairplot(df1[df1['Win'] > 5], 
             vars = ['Off.TDs', 'Touchdowns', 'PAT'], diag_kind = 'kde', 
             plot_kws = {'alpha': 0.6, 's': 80, 'edgecolor': 'k'},
             size = 4);
# Title 
plt.suptitle('Pair Plot of influential factors for winning', 
             size = 28);
```
